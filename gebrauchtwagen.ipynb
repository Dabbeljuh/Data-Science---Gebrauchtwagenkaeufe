## Abschlussprojekt: Gebrauchtwagenkäufe

<p><strong>Szenario:</strong> Du arbeitest als Data Scientist für einen US-amerikanischen Gebrauchtwagenhändler. Der Händler kauft gebrauchte Autos günstig in Onlineauktionen und von anderen Autoverkäufern, um sie dann auf der eigenen Plattform gewinnbringend weiterzuverkaufen. Es ist nicht immer einfach zu erkennen, ob sich der Kauf eines Gebrauchtwagens lohnt: Eine der größten Herausforderungen bei Gebrauchtwagenauktionen ist das Risiko, dass ein Auto solch schwerwiegende Probleme hat, die verhindern, dass es an Kunden weiterverkauft werden kann. Hierbei handelt es sich um sogenannte "Montagsautos" - also Autos, die von Hause aus erhebliche Mängel aufgrund von Produktionsfehlern aufweisen, welche die Sicherheit, die Verwendung oder den Wert dieses Autos erheblich beeinträchtigen und gleichzeitig nicht in einer angemessenen Anzahl an Reparaturen oder innerhalb eines bestimmten Zeitraums behoben werden können. Der Kunde hat in diesem Fall das Recht, sich den Kaufpreis zurückerstatten zu lassen. Neben den Anschaffungskosten führt der Fehleinkauf von solchen "Montagsautos" also zu erheblichen Folgekosten, wie z.B. der Einlagerung und Reparatur des Wagens, welche in Verlusten beim Weiterverkauf des Fahrzeugs resultieren können.</p>
<p>Deshalb ist es deiner Chefin wichtig, möglichst viele Fehlkäufe von "Montagsautos" auszuschließen. Um die Einkäufer im Unternehmen bei der riesigen Anzahl an Angeboten zu entlasten, sollst du ein Modell entwickeln, welches vorhersagt, ob ein Angebot ein Fehlkauf im Sinne eines Montagsautos wäre. Das darf allerdings nicht dazu führen, dass zu viele gute Käufe ausgeschlossen werden. Genauere Angaben zu den Kosten und Gewinnen der jeweiligen Käufe erhältst du für die Entwicklung des Prototyps noch nicht.</p>

<p>Welches Problem soll das Modell lösen?</p>
<p>Das Modell soll vorhersagen, ob ein Fahrzeug, das bei einer Auktion gekauft wurde, ein "Montagsauto" ist, also einen erheblichen Defekt aufweist, der es unverkäuflich macht. 
Das Hauptziel besteht darin, die Anzahl fehlerhafter Käufe zu minimieren, um finanzielle Verluste zu vermeiden.</p>
<p>Welcher Art ist das Problem (z.B. Klassifikation, Regression, Clustering ...)?</p>
<p>Das Problem ist eine Klassifikationsaufgabe. Es geht darum, anhand der gegebenen Merkmale zu bestimmen, ob ein Fahrzeug ein "Montagsauto" ist oder nicht.</p>
<p>Wie sähe eine Anwendung aus, die dein Modell benutzt?</p>
<p>Eine mögliche Anwendung könnte darin bestehen, dass der Gebrauchtwagenhändler das Modell verwendet, um potenzielle Fahrzeugkäufe bei Auktionen zu bewerten. 
Bei neuen Angeboten könnte das Modell automatisch eine Vorhersage treffen, ob es sich lohnt, das Fahrzeug zu kaufen oder nicht. 
Auf diese Weise können potenzielle Verluste durch den Kauf fehlerhafter Fahrzeuge reduziert werden.</p>
<p>Welche Anforderungen stellt der Auftraggeber an dein Modell?</p>
<p>Der Auftraggeber möchte, dass das Modell möglichst viele fehlerhafte Käufe von "Montagsautos" ausschließt, jedoch ohne zu viele gute Käufe auszuschließen. 
Das bedeutet, dass das Modell eine hohe Genauigkeit bei der Identifizierung von "Montagsautos" haben sollte, aber auch eine angemessene Anzahl guter Käufe zulassen sollte.</p>



<p>Da das Ziel darin besteht, die Anzahl der Fehleinkäufe von Montagsautos zu minimieren, ist es wichtig, sowohl Precision als auch Recall zu berücksichtigen. Ein Modell mit einer guten Balance zwischen Precision und Recall wäre wünschenswert, um möglichst viele tatsächliche Fehleinkäufe zu identifizieren und gleichzeitig die Anzahl der falsch positiven Vorhersagen zu minimieren.
F1-Score als harmonisches Mittelwert zwischen Precision und Recall verwendet werden, um eine einzige Metrik zu haben, die beide Aspekte berücksichtigt.</p>

```python
drop_cols = {}

def cols_to_drop(col, comment):
    """
    This function is used to add a column and its corresponding comment to a dictionary called 'drop_cols'.
    If the given column already exists in the 'drop_cols' dictionary, its comment will be updated.
    """
    if col not in drop_cols:
        drop_cols[col] = comment
    return drop_cols

def show_cols_to_drop():
    """This function iterates over the 'drop_cols' dictionary and prints each column name along with its comment.
    """
    print('columns to drop: ')
    for col, comment in drop_cols.items():
        print(f'{col}: {comment}')
```
